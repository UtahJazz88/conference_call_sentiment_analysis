```python
# CAR = (原始回报 - 市场加权回报) * 100
#Cumulative abnormal returns in the three-day window around the earnings conference call date, multiplied by 100. The abnormal return equals raw return minus value-weighted market return
df['CAR'] = (df['car']) * 100
```

```python
#EARN
#Income before extraordinary items in a quarter scaled by the book value of assets at the beginning of the quarter
df['Earn'] = df['ibq'] / df['begin_of_quarter_assets']
```

```python
#UE
#Actual quarterly earnings per share (EPS) minus analyst consensus EPS forecast issued immediately before the earnings announcement date, scaled by the share price at the end of the fiscal quarter and multiplied by 100. 
#Actual EPS and analyst consensus EPS forecasts are from I/B/E/S
# 假设df['analyst_consensus_eps']是从I/B/E/S数据库获取的分析师共识EPS
#這個還沒找到
df['UE'] = (df['epspxq'] - df['analyst_consensus_eps']) / df['prccq'] * 100
```

```python
#Accrual
'''
Net income（niq）
minus 
cash flow from operations during the quarter (oancfy:Operating Activities - Net Cash Flow)
scaled by 
the book value of assets at the beginning of the quarter
'''
df['Accruals'] = (df['niq'] - df['oancfy']) / df['begin_of_quarter_assets']
```

```python
#EarnVol
'''
Standard deviation of the return on assets during the prior 16 quarters, 
with return on assets calculated as quarterly earnings 
scaled by total assets at the beginning of that quarter'''
df['roa'] = df['niq'] / df['begin_of_quarter_assets']
df['EarnVol'] = df.groupby('gvkey')['roa'].transform(lambda x: x.rolling(window=16, min_periods=1).std())
```

```python
#MtoB
#Market value of equity plus book value of total liabilities scaled by book value of total assets at the end of the quarter
df['MtoB'] = (df['mkvaltq'] + df['ltq']) / df['atq']
```

```python
import json
import ast
def calculate_average_sentiment(speech_data_str):
    # 使用ast.literal_eval安全地解析字符串
    try:
        speech_data = ast.literal_eval(speech_data_str)
    except ValueError:
        return {}

    label_to_score = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
    sentiment_scores = {}

    for speaker, speeches in speech_data.items():
        total_score = sum(label_to_score[speech['label']] for speech in speeches)
        average_score = total_score / len(speeches) if speeches else 0
        sentiment_scores[speaker] = average_score

    return sentiment_scores


# 应用函数计算
df['Pre Sentiment Scores'] = df['PresentationPreSpeakerSentimentResults'].apply(calculate_average_sentiment)
# 应用函数计算

# 输出计算结果
print(df['Pre Sentiment Scores'])
```

```python
df['QA Sentiment Scores'] = df['QandAPreSpeakerSentimentResults'].apply(calculate_average_sentiment)
```

```python
def calculate_difference(row):
    pre_scores = row['Pre Sentiment Scores']
    qa_scores = row['QA Sentiment Scores']
    
    # 获取所有唯一的发言者
    all_speakers = set(pre_scores.keys()).union(set(qa_scores.keys()))
    
    # 计算差异
    score_differences = {}
    for speaker in all_speakers:
        pre_score = pre_scores.get(speaker, 0)  # 如果发言者不存在于pre中，返回0
        qa_score = qa_scores.get(speaker, 0)   # 如果发言者不存在于QA中，返回0
        score_differences[speaker] = qa_score - pre_score
    
    return score_differences


df['Sentiment Score Differences'] = df.apply(calculate_difference, axis=1)

```
