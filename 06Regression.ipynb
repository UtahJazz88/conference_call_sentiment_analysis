```python
# CAR = (原始回报 - 市场加权回报) * 100
#Cumulative abnormal returns in the three-day window around the earnings conference call date, multiplied by 100. The abnormal return equals raw return minus value-weighted market return
df['CAR'] = (df['car']) * 100
```

```python
#EARN
#Income before extraordinary items in a quarter scaled by the book value of assets at the beginning of the quarter
df['Earn'] = df['ibq'] / df['begin_of_quarter_assets']
```

```python
#UE
#Actual quarterly earnings per share (EPS) minus analyst consensus EPS forecast issued immediately before the earnings announcement date, scaled by the share price at the end of the fiscal quarter and multiplied by 100. 
#Actual EPS and analyst consensus EPS forecasts are from I/B/E/S
# 假设df['analyst_consensus_eps']是从I/B/E/S数据库获取的分析师共识EPS
#這個還沒找到
df['UE'] = (df['epspxq'] - df['analyst_consensus_eps']) / df['prccq'] * 100
```

```python
import json
import ast
def calculate_average_sentiment(speech_data_str):
    # 使用ast.literal_eval安全地解析字符串
    try:
        speech_data = ast.literal_eval(speech_data_str)
    except ValueError:
        return {}

    label_to_score = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
    sentiment_scores = {}

    for speaker, speeches in speech_data.items():
        total_score = sum(label_to_score[speech['label']] for speech in speeches)
        average_score = total_score / len(speeches) if speeches else 0
        sentiment_scores[speaker] = average_score

    return sentiment_scores


# 应用函数计算
df['Pre Sentiment Scores'] = df['PresentationPreSpeakerSentimentResults'].apply(calculate_average_sentiment)
# 应用函数计算

# 输出计算结果
print(df['Pre Sentiment Scores'])
```

```python
#d第二種方法算sentiment
import json
import ast
def calculate_average_weight_sentiment(speech_data_str):
    # 使用ast.literal_eval安全地解析字符串
    try:
        speech_data = ast.literal_eval(speech_data_str)
    except ValueError:
        return {}

    label_to_score = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
    sentiment_scores = {}

    for speaker, speeches in speech_data.items():
        total_score = 0
        total_weight = 0  # 总权重，用于加权平均

        for speech in speeches:
            label_score = label_to_score[speech['label']]
            score = speech['score']  # 情绪概率
            total_score += label_score * score  # 加权得分
            total_weight += score  # 叠加每个标签的分数，作为权重

        # 计算加权平均得分
        average_score = total_score / total_weight if total_weight > 0 else 0
        sentiment_scores[speaker] = average_score

    return sentiment_scores


# 应用函数计算
df['Pre Weight Sentiment Scores'] = df['PresentationPreSpeakerSentimentResults'].apply(calculate_average_weight_sentiment)
# 应用函数计算

# 输出计算结果
print(df['Pre Weight Sentiment Scores'])
```
