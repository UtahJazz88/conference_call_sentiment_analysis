```python
import pandas as pd
import re

def extract_and_transform_quarter(title):
    # 使用正则表达式匹配格式如 "Q1 2023"
    match = re.search(r'Q[1-4]\s(\d{4})', title)
    if match:
        # 提取季度和年份
        quarter = match.group(0)
        # 转换格式为 "2023Q1"
        year, q = match.groups()[0], quarter[0:2]
        return f"{year}{q}"
    else:
        return None  

    
df_sandp500_ECT_data['quarter_transformed'] = df_sandp500_ECT_data['title'].apply(extract_and_transform_quarter)
```

```python
'''
TICKER: 股票交易代码，用于在股票市场上唯一标识一家公司。
CUSIP: Committee on Uniform Securities Identification Procedures的缩写，是一个9位的证券识别代码，用于在美国和加拿大识别证券。
OFTIC: 通常是公司的官方股票交易代码。
MEASURE: 测量指标的类型，这里是EPS，即每股收益。
FPI: Forecast Period Indicator（预测期指示符），一个数字表示这个预测是关于哪个时间段的。
NUMEST: Number of Estimates（估计数量），表示有多少个不同的分析师或数据源提供了这个预测。
MEANEST: Mean Estimate（平均估计），这是所有提供的预测的平均值。
USFIRM: 表示这个数据是否来自于美国的公司。通常1表示是，0表示否。
FPEDATS: Forecast Period End Date（预测期结束日期），这个日期表示预测的财务指标是哪个期间结束的日期。
'''
wrds_IBES
```
