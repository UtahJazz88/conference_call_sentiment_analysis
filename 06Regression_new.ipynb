```python
# CAR = (原始回报 - 市场加权回报) * 100
#Cumulative abnormal returns in the three-day window around the earnings conference call date, multiplied by 100. The abnormal return equals raw return minus value-weighted market return
df['CAR'] = (df['car']) * 100
```

```python
#EARN
#Income before extraordinary items in a quarter scaled by the book value of assets at the beginning of the quarter
df['Earn'] = df['ibq'] / df['begin_of_quarter_assets']
```

```python
#UE
'''第一種方法算'''
#Actual quarterly earnings per share (EPS) minus analyst consensus EPS forecast issued immediately before the earnings announcement date, scaled by the share price at the end of the fiscal quarter and multiplied by 100. 
#Actual EPS and analyst consensus EPS forecasts are from I/B/E/S
# 假设df['analyst_consensus_eps']是从I/B/E/S数据库获取的分析师共识EPS
#這個還沒找到
# df['UE'] = (df['epspxq'] - df['analyst_consensus_eps']) / df['prccq'] * 100

'''第二種方法算'''
# 按照gvkey和datadate排序
df = df.sort_values(by=['gvkey', 'datadate'])
# 計算上一期盈餘
df['previous_niq'] = df.groupby('gvkey')['niq'].shift(1)

df['UE'] = (df['niq'] - df['previous_niq']) / df['prccq']
```

```python
#Accrual
'''
Net income（niq）
minus 
cash flow from operations during the quarter (oancfy:Operating Activities - Net Cash Flow)
scaled by 
the book value of assets at the beginning of the quarter
'''
df['Accruals'] = (df['niq'] - df['oancfy']) / df['begin_of_quarter_assets']
```

```python
#EarnVol

'''
Standard deviation of the return on assets during the prior 16 quarters, 
with return on assets calculated as quarterly earnings 
scaled by total assets at the beginning of that quarter'''
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# 计算 ROA
df['roa'] = df['niq'] / df['begin_of_quarter_assets']

# 使用 groupby 和 rolling 计算 EarnVol
earn_vol = df.groupby('gvkey')['roa'].rolling(window=16, min_periods=1).std().reset_index(level=0, drop=True)

# 确保 EarnVol 列与 DataFrame 的长度匹配
df['EarnVol'] = earn_vol

# 使用 IterativeImputer 填补 NaN 值
imputer = IterativeImputer(max_iter=10, random_state=0)
df[['EarnVol']] = imputer.fit_transform(df[['EarnVol']])


```

```python
import json
import ast
def calculate_average_sentiment(speech_data_str):
    # 使用ast.literal_eval安全地解析字符串
    try:
        speech_data = ast.literal_eval(speech_data_str)
    except ValueError:
        return {}

    label_to_score = {'Positive': 1, 'Neutral': 0, 'Negative': -1}
    sentiment_scores = {}

    for speaker, speeches in speech_data.items():
        total_score = sum(label_to_score[speech['label']] for speech in speeches)
        average_score = total_score / len(speeches) if speeches else 0
        sentiment_scores[speaker] = average_score

    return sentiment_scores


# 应用函数计算
df['Pre Sentiment Scores'] = df['PresentationPreSpeakerSentimentResults'].apply(calculate_average_sentiment)
df['QA Sentiment Scores'] = df['QandAPreSpeakerSentimentResults'].apply(calculate_average_sentiment)
df['QA Analyst Sentiment Scores'] = df['QandAPreAnalystSentimentResults'].apply(calculate_average_sentiment)

# 输出计算结果
print(df['Pre Sentiment Scores'])
```

```python
def calculate_difference(row):
    pre_scores = row['Pre Sentiment Scores']
    qa_scores = row['QA Sentiment Scores']
    
    # 获取所有唯一的发言者
    all_speakers = set(pre_scores.keys()).union(set(qa_scores.keys()))
    
    # 计算差异
    score_differences = {}
    for speaker in all_speakers:
        pre_score = pre_scores.get(speaker, 0)  # 如果发言者不存在于pre中，返回0
        qa_score = qa_scores.get(speaker, 0)   # 如果发言者不存在于QA中，返回0
        score_differences[speaker] = qa_score - pre_score
    
    return score_differences


df['Sentiment Score Differences'] = df.apply(calculate_difference, axis=1)

```

```python
#每一家公司每一家公司跑回歸
def calculate_average_scores(scores):
    if scores:
        return sum(scores.values()) / len(scores)
    return 0

# 假设 'Sentiment Score Differences' 是正确的列名，且它包含字典
df['Average Sentiment Difference'] = df['Sentiment Score Differences'].apply(calculate_average_scores)
#爲了看數據而已 回歸不用
df['Pre Average Sentiment']  = df['Pre Sentiment Scores'].apply(calculate_average_scores)
df['QA Average Sentiment']  = df['QA Sentiment Scores'].apply(calculate_average_scores)
df['QA Anaylst Average Sentiment']  = df['QA Analyst Sentiment Scores'].apply(calculate_average_scores)
```

```python
# 函数：提取每个演讲者的句子总数
def count_sentences(speech_data_str):
    # 使用 ast.literal_eval 安全地解析字符串
    try:
        speech_data = ast.literal_eval(speech_data_str)
    except ValueError:
        return {}

    sentence_counts = {}

    for speaker, speech in speech_data.items():
        # 按句号分割句子并计算数量
        sentences = speech.split('.')
        sentence_counts[speaker] = len([s for s in sentences if s.strip()])  # 去除空句子

    return sentence_counts

df['PresentationPreSpeaker_Sentence_Counts'] = df['PresentationPreSpeaker'].apply(count_sentences)
df['QandAPreASpeaker_Sentence_Counts'] = df['QandAPreASpeaker'].apply(count_sentences)
df['QandAPreQAnalyst_Sentence_Counts'] = df['QandAPreQAnalyst'].apply(count_sentences)
```

```python
# 计算总句子数和平均句子数的函数
def calculate_total_and_avg_sentences(df, column):
    # 将每个演讲者的句子数展开为多个列
    sentence_counts_df = df[column].apply(pd.Series)
    
    # 计算每个演讲者的总句子数
    total_sentence_counts = sentence_counts_df.sum()
    
    # 计算所有演讲者的总句子数
    total_sentences = total_sentence_counts.sum()
    
    # 计算每个演讲者的平均句子数
    avg_sentences = sentence_counts_df.mean()
    
    return total_sentence_counts, total_sentences, avg_sentences

# 计算结果并添加到数据框中
for column in ['PresentationPreSpeaker_Sentence_Counts', 'QandAPreASpeaker_Sentence_Counts', 'QandAPreQAnalyst_Sentence_Counts']:
    total_sentence_counts, total_sentences, avg_sentences = calculate_total_and_avg_sentences(df, column)
    
    # 添加总句子数和平均句子数列
    df[f'{column}_Total'] = df[column].apply(lambda x: sum(x.values()))
    df[f'{column}_Avg'] = df[column].apply(lambda x: sum(x.values()) / len(x) if len(x) > 0 else 0)

```

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize

# 下载 NLTK 资源
nltk.download('punkt')

def count_words(text):
    """计算文本中的词总数"""
    words = word_tokenize(text)
    return len(words)

def process_word_count_columns(df, column_name):
    df[column_name + '_WordCount'] = [None] * len(df)
    
    for index, row in df.iterrows():
        speakers_dict_str = row[column_name]
        
        # 确保字符串正确转换为字典
        try:
            speakers_dict = eval(speakers_dict_str)
        except:
            print(f"Error parsing string to dict at index {index}: {speakers_dict_str}")
            continue
        
        all_speakers_word_counts = {}
        
        for speaker, text in speakers_dict.items():
            word_count = count_words(text)
            all_speakers_word_counts[speaker] = word_count
        
        df.at[index, column_name + '_WordCount'] = all_speakers_word_counts

    return df


# 处理DataFrame中的指定列以计算词总数
df_non_financials = process_word_count_columns(df_non_financials, 'PresentationPreSpeaker')
df_non_financials = process_word_count_columns(df_non_financials, 'QandAPreASpeaker')
df_non_financials = process_word_count_columns(df_non_financials, 'QandAPreQAnalyst')
import ast
# 检查并转换数据为字典格式
def convert_to_dict(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except (ValueError, SyntaxError):
            print(f"无法解析为字典: {x}")
            return {}
    return x if isinstance(x, dict) else {}

# 转换所有相关列
df_non_financials['PresentationPreSpeaker_WordCount'] = df_non_financials['PresentationPreSpeaker_WordCount'].apply(convert_to_dict)
df_non_financials['QandAPreASpeaker_WordCount'] = df_non_financials['QandAPreASpeaker_WordCount'].apply(convert_to_dict)
df_non_financials['QandAPreQAnalyst_WordCount'] = df_non_financials['QandAPreQAnalyst_WordCount'].apply(convert_to_dict)


# 调试总字数计算
def calculate_total_word_count(word_count_dict):
    if isinstance(word_count_dict, dict):
        total = sum(word_count_dict.values())
        print(f"字典: {word_count_dict} -> 总字数: {total}")
        return total
    print(f"非字典数据: {word_count_dict}")
    return 0

# 调试平均字数计算
def calculate_avg_word_count(word_count_dict):
    if isinstance(word_count_dict, dict) and len(word_count_dict) > 0:
        avg = sum(word_count_dict.values()) / len(word_count_dict)
        print(f"字典: {word_count_dict} -> 平均字数: {avg}")
        return avg
    print(f"非字典数据或空字典: {word_count_dict}")
    return 0

# 计算总字数和平均字数
df_non_financials['PresentationPreSpeaker_WordCount_Total'] = df_non_financials['PresentationPreSpeaker_WordCount'].apply(calculate_total_word_count)
df_non_financials['QandAPreASpeaker_WordCount_Total'] = df_non_financials['QandAPreASpeaker_WordCount'].apply(calculate_total_word_count)
df_non_financials['QandAPreQAnalyst_WordCount_Total'] = df_non_financials['QandAPreQAnalyst_WordCount'].apply(calculate_total_word_count)

df_non_financials['PresentationPreSpeaker_WordCount_Avg'] = df_non_financials['PresentationPreSpeaker_WordCount'].apply(calculate_avg_word_count)
df_non_financials['QandAPreASpeaker_WordCount_Avg'] = df_non_financials['QandAPreASpeaker_WordCount'].apply(calculate_avg_word_count)
df_non_financials['QandAPreQAnalyst_WordCount_Avg'] = df_non_financials['QandAPreQAnalyst_WordCount'].apply(calculate_avg_word_count)
```

```python
import ast

def parse_string_to_list(row):
    try:
        # 尝试将字符串解析为列表
        return ast.literal_eval(row)
    except:
        # 如果解析失败，返回空列表
        return []

# 应用这个转换到 CallParticipants 列
df_CEO['CallParticipants'] = df_CEO['CallParticipants'].apply(parse_string_to_list)

```

```python
def find_ceo_in_qanda(row):
    ceo_name = row['CEO'][0] if row['CEO'] and len(row['CEO']) > 0 else None
    if not ceo_name:
        return None  # 如果没有 CEO 名字，则不进行检查

    q_and_a = row['QandA']
    participants = row['CallParticipants']

    # 初始化记录提到 CEO 的参与者列表
    mentioned_by = []

    if len(participants) > 0:
        for i in range(len(participants) - 1):
            try:
                # 查找第一个参与者在 QandA 中的第一次出现
                start_pos = q_and_a.index(participants[i])
                # 查找第二个参与者在 QandA 中的第一次出现
                end_pos = q_and_a.index(participants[i+1], start_pos)
                # 检查两者之间的文本是否包含 CEO 名字
                if ceo_name in q_and_a[start_pos:end_pos]:
                    mentioned_by.append(participants[i])
            except ValueError:
                continue  # 如果找不到这个参与者的名字，跳过

        # 检查最后一个参与者
        try:
            last_start_pos = q_and_a.index(participants[-1])
            if ceo_name in q_and_a[last_start_pos:]:
                mentioned_by.append(participants[-1])
        except ValueError:
            pass  # 如果找不到最后一个参与者的名字，跳过

    return mentioned_by if mentioned_by else None

# 应用这个函数到 DataFrame
df_CEO['AskertoCEO'] = df_CEO.apply(find_ceo_in_qanda, axis=1)


```

```python
def extract_Asker_sentiment(row):
    ceo_dict = row['QA Analyst Sentiment Scores']
    askers = row['AskertoCEO']
    if isinstance(askers, list):  # 检查 askers 是否是列表
        total_score = 0
        count = 0
        for asker in askers:
            # 在情绪分数字典中查找每个问问题的人的分数
            if asker in ceo_dict:
                total_score += ceo_dict[asker]
                count += 1
        if count > 0:
            return total_score / count  # 计算平均情绪分数
    return None  # 如果没有有效的问问题的人或没有情绪分数，返回 None

# 在 'Asker_Sentiment' 列中应用函数，提取提问者的平均情绪分数
df_CEO['Asker_Sentiment'] = df_CEO.apply(extract_Asker_sentiment, axis=1)
df_CEO['Asker_Sentiment'] = df_CEO['Asker_Sentiment'].fillna(0)

print(df_CEO['Asker_Sentiment'])

```
